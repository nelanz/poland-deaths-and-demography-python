```python
def get_df (page_url):
    response_API = requests.get(page_url)
    population_summary = json.loads(response_API.text)
    rows = []

    for data in population_summary['results']:
        for value in data['values']:
            row = {'id': data['id'],
                'name': data['name'],
                'year': value['year'],
                'value': value['val']}
            rows.append(row)
    df = pd.DataFrame(rows)
    return df
```

```python
df_0 = get_df(page_0)
df_1 = get_df(page_1)
df_2 = get_df(page_2)
df_3 = get_df(page_3)
```
