```python
poland = gpd.read_file('https://www.gis-support.pl/downloads/2022/powiaty.zip')
```

```python
poland['powiat_nazwa'] = poland['JPT_NAZWA_'].str.replace('powiat ', '')
```

```python
poland_info = poland[['JPT_POWIER', 'JPT_KOD_JE', 'WERSJA_OD', 'WAZNY_OD', 'SHAPE_LENG', 'SHAPE_AREA', 'geometry', 'powiat_nazwa']]
```

```python
ALL_DATA = pd.read_csv('../results/ALL_DATA_AGE_EUROSTAT.csv')
```

```python
ALL_DATA_Y = pd.read_csv('../results/all_poviats_stats.csv')
```

```python
ALL_DATA_Y
```

```python
poviats_data_Y = ALL_DATA_Y.query('powiat_numer > 32')
```

```python
poviats_data_coord_Y = pd.merge(poviats_data_Y, poland_info, on="powiat_nazwa")
```

```python
poviats_data_coord_Y
```

```python
age_group_select = widgets.Dropdown(
    options=list(poviats_data_coord_Y.wiek_eurostat.unique()),
    value='TOTAL',
    description='Age group:',
    disabled=False,
)
```

```python
# Create the year slider widget
year_slider = widgets.IntSlider(min=poviats_data_coord_Y['YEAR'].min(), max=poviats_data_coord_Y['YEAR'].max(), step=1, value=poviats_data_coord_Y['YEAR'].min())
```

```python
week_slider = widgets.IntSlider(min=1, max=52, step=1, value=1, description='T')
```

### Yearly total number of deaths in poviats

```python
# Create a function to update the plot based on the selected year
def update_plot(year, age_group):
    year_data = gpd.GeoDataFrame(poviats_data_coord_Y.query(f'YEAR == {year} & wiek_eurostat == "{age_group}"'))
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 6))
    year_data.plot(ax=ax, column='sum', cmap='viridis', legend=True)
    ax.set_axis_off()
    ax.set_title(f'Total number of deaths for age group {age_group} in the year {year}', fontsize=16)
    plt.show()


# Link the slider widget to the update function
widgets.interact(update_plot, year=year_slider, age_group=age_group_select)
```

### Weekly total number of deaths in poviats

```python
ALL_DATA_W = pd.read_csv('../results/ALL_DATA_TS.csv')
```

```python
ALL_DATA_W
```

```python
poviats_data_W = ALL_DATA_W.query('powiat_numer > 32')
```

```python
poviats_data_coord_W = pd.merge(poviats_data_W, poland_info, on="powiat_nazwa")
```

```python
def update_plot_weekly(year, week, age_group):
    week_formatted = f'{week:02d}'
    week_data = gpd.GeoDataFrame(poviats_data_coord_W.query(f'YEAR == {year} & week == "T{week_formatted}" & wiek_eurostat == "{age_group}"'))
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 6))
    week_data.plot(ax=ax, column='count', cmap='viridis', legend=True)
    ax.set_axis_off()
    ax.set_title(f'Total number of deaths for age group {age_group} in the week {week} of the year {year}', fontsize=16)
    plt.show()


# Link the slider widget to the update function
widgets.interact(update_plot_weekly, year=year_slider, week = week_slider, age_group=age_group_select)

```
